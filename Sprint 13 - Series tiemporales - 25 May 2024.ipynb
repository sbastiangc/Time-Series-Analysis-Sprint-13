## Formación

### Creación de Características

```python
# eliminó las columnas creadas anteriormente
df = df.drop(['mean','std'], axis=1)

#creo las columnas para extraer cada componente de la fecha mes, dow, dia y hora
df['month'] = df.index.month
df['day_of_week'] = df.index.dayofweek
df['day'] = df.index.day
df['hour'] = df.index.hour

# verifico que las columnas se crearan correctamente
df.head()
```

```python
# creo 4 columnas de lag y uso shift para crear el desfase
for lag in range (1,5):
    df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
    
df.head()
```

```python
#creo el promedio móvil para las últimas 4 horas
df['rolling_mean'] = df['num_orders'].shift().rolling(4).mean()
df.head()
```

```python
# ya que tenemos nuevos datos con N/A los elimnare
df.dropna(inplace = True)
```

**Observaciones:**

- Se crearon las columnas de características `month`, `day_of_week`, `day`,`hour` extrayendo los datos del índice de cada registro.
- Se crearon las columnas `lag_1`, `lag_2`, `lag_3`, `lag_4`, con el uso de sifth.rolling()
-  Se creo `rolling_mean` para las últimas cada 4 horas.

### Segmentación de datos

```python
#uso train_test_split para dividir los datos en entrenamiento y prueba con 10% - suffle= False
train, test = train_test_split(df, shuffle=False, test_size=0.1)  

#uso train_test_split para dividir los datos en entrenamiento y validación con 10% - suffle= False
train, valid = train_test_split(train, shuffle=False, test_size=0.1111) 
```

```python
#creo features y target en para entrenamiento
features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']

#creo features y target en para validación
features_valid = valid.drop(['num_orders'], axis=1)
target_valid = valid['num_orders']

#creo features y target en para prueba
features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']
```

```python
# verifico el tamaño de los segmentos de prueba
print('features_train:', features_train.shape)
print('target_train:', target_train.shape)

print('')
print('features_valid', features_valid.shape)
print('target_valid', target_valid.shape)

print('')
print('features_test', features_test.shape)
print('target_test', target_test.shape)
```

**Observaciones:**
- Los tamaños de cada segmento corresponden a **80% para entrenamiento, 10% validación y 10% prueba.**

### Entrenamiento del modelo

#### Regresión lineal

```python
# defino el modelo de regresión lineal
model = LinearRegression()

#entreno los datos y hago las predicciones con los datos de validación
model.fit(features_train, target_train)
pred_valid = model.predict(features_valid)

#calculo RMSE
lr_rmse = mean_squared_error(target_valid, pred_valid)**0.5

print('Regresión lineal:')
print("RMSE:", lr_rmse)
```

#### Regresor de árbol

```python
# usaré GridSearchCV para invetigar los hiperparametros

# defino el modelo y agrego como validación cruzada el metodo timeseriessplit
model = DecisionTreeRegressor(random_state=12345)
cv = TimeSeriesSplit()

# creo un diccionario de hiperparametros a combinar
params = {
    'max_depth': [5, 10, 15, 20, 25, 30],
    'min_samples_split': [2, 5, 7, 10, 12],
    'min_samples_leaf': [1, 2, 4, 8, 10]
}

# creo el objeto grid_serach y defino los hiperparametros: scoring = RMSE 
grid_search = GridSearchCV(estimator= model,       
                           param_grid= params, 
                           scoring='neg_mean_squared_error', 
                           cv=cv)

# entreno el objeto grid_search con datos de entrenamiento
grid_search.fit(features_train, target_train)

# guardo el resultado del mejor modelo con el atributo best_estimator_
# guardo la predicción con los datos de validación
best_model = grid_search.best_estimator_
pred_valid = best_model.predict(features_valid)

# calculo el RMSE para el mejor modelo
dtr_rmse = mean_squared_error(target_valid, pred_valid)**0.5

# imprimo los hiperparametros utilizados y el RMSE
print('Regresor de árbol:')
print("Mejores hiperparametros:", grid_search.best_params_)
print("RMSE:", dtr_rmse)
```

#### Regresor de bosque aleatorio

**Nota:** en la investigación se usaron más posibles hiperparámetros, se conservó solo la mejor combinación para optimizar el rendimiento del notebook, ya que esta celda tardaba ~2min en ejecutar.

```python
# usaré GridSearchCV para investigar los hiperparametros

# defino el modelo y agrego como validación cruzada el metodo timeseriessplit
model = RandomForestRegressor(random_state=12345)
cv = TimeSeriesSplit()

# creo un diccionario de hiperparametros a combinar
params = {
    'n_estimators': [200],
    'max_depth': [25],
    'min_samples_leaf': [1]}


# creo el objeto grid_serach y defino los hiperparametros: scoring = RMSE 
grid_search = GridSearchCV(estimator= model,       
                           param_grid= params, 
                           scoring='neg_mean_squared_error', 
                           cv=cv)

grid_search.fit(features_train, target_train)
best_model = grid_search.best_estimator_
pred_valid = best_model.predict(features_valid)

# calculo el RMSE para el mejor modelo
rf_rmse = mean_squared_error(target_valid, pred_valid)**0.5

# imprimo los hiperparametros utilizados y el RMSE
print('Regresor Bosque aleatorio:')
print("Mejores hiperparametros:", grid_search.best_params_)
print("RMSE:", rf_rmse)
```
